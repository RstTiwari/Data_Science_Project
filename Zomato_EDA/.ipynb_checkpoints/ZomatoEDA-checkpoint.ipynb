```python
df_country_code = pd.read_excel("Country-Code.xlsx")
```

```python
df_country_code.head()
```

```python
final_dataset = pd.merge(df , df_country_code , on = "Country Code", how = "left" )
final_dataset.head(2)
```

```python
country_name = final_dataset["Country"].value_counts().index
country_count  = final_dataset["Country"].value_counts().values
print(country_count,country_name)
```

```python
plt.pie(country_count[:3] , labels = country_name[:3] ,autopct = "%1.2f%%")
```

```python
#Conclusions are 
# 1) Zomato is having maximum numbers of Orders from india 
# 2) Zomatao revenue comes fro india followed by United states and india 
```

```python
final_dataset.columns
```

```python
rating  = final_dataset.groupby(["Aggregate rating", "Rating color" , "Rating text" ]).size().reset_index().rename(columns= {0:"Rating Count"})
rating
```

```python
# LETS gROUP BY IT ON THE RATING COLOR
ratings_by = rating.groupby(["Rating color","Rating Count"]).size()
```

```python
#Observations from the above table is clear
## Avarage rating - 0.0 => Not Rateed
## Avarage rating 1.8 - 2.4 => Poor Rated 
## Avarage rating 2.5 - 3.4  => Avarage Rated
## Avarage raitng 3.5 - 4.4 => Good Rated
## Avarage rating 4.5 - 4.9 = > Excellent Rated
```

```python
#now Will plot the graph for the given Data opf rating
plt.rcParams["figure.figsize"] =  (15,5)
sns.barplot(rating , x= "Aggregate rating" , y = "Rating Count" ,hue = "Rating color",palette=["blue","red","orange","yellow","green","green"])
```
